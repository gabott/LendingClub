# Loan Case Study

### 1) Identify the risky loan applicants to  reduce the amount of credit loss.
### 2) Identify the variables which are strong indicators of default
### 3) Identify the consumer attributes and loan attributes influence the tendency of default.

# Observation:
There are 15% default rate in the data

## The Probability of default cases are

1) When applicant has house_ownership as "rent"
2) When applicant is from grade B at the 5 level
3) When applicant has purpose of debt clearence
4) When applicant term is 36 months
5) Most of the default cases when the aplicant details are not verified
6) When the fund amount by investor is between 5k to 10k
7) When the loan amount between 5k to 10k
8) When the public records are 0
9) When the annual income types between 31k to 50k 
10) When the dti is between 12 to 18
11) When montly installment between 145 to 274
12) When the interest rate is 13% to 17%
13)  When the application of the length 10

## The approach for analysis is divided into four main parts:
### a. Data understanding from the csv files provided.
### b. Data cleaning (cleaning missing values, removing redundant columns etc.). 
### c. Data Analysis (Univariate, Bivariate).
### d. Recommendations.
